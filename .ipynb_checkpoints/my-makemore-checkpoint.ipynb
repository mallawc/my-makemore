### Bigram
在二元语言模型中，每次只处理两个字符。因此设置一个窗口，只看给定的一个字符，并试图预测序列中的下一个字符。我们只建模这种小范围的局部结构，忽略可能拥有更多信息的事实，所以这是一个非常简单且弱的语言模型，但这是一个很好的学习起点：

python很有用的一个特性：利用zip函数接收两个迭代器（iterators），将它们配对，然后创建一个迭代器遍历它们连续项的元组

```python
b = {}
for w in words[:3]:
    #引入自定义的头尾标记，w包装成list：
    chs = ['<S>'] + list(w) + ['<E>']
    for ch1, ch2 in zip(chs, chs[1:]):
        bigram = (ch1, ch2)
        # 通过字典get方法统计二元组出现次数（+1）
        b[bigram] = b.get(bigram,0) + 1
        print(ch1,ch2)
```

我们留意到，e-m、m-m、m-a结对都表示了两个字符的关系，但是对于“以e开头”、“以a结尾”的语义没有表示清楚，所以引入一个自定义的特殊起始（和结尾）标记（special token）

把w包装（wrapping）成列表的原因是，w是一个字符串

为了学习哪些字符可能跟随其他字符的统计信息，在大规模语言模型中，最简单的方法就是通过计数的方法来实现，即统计这些组合在训练集中出现的频率。
所以我们需要某种字典，来维护这些二元组中每一个的计数，用dictionary b来映射（map）这些二元组（bigram）。二元组由ch1、ch2组成的元组

```python
b = {}
for w in words:
    chs = ['<S>'] + list(w) + ['<E>']
    for ch1, ch2 in zip(chs, chs[1:]):
        bigram = (ch1, ch2)
        b[bigram] = b.get(bigram,0) + 1
```

```python
#默认下，sort是对“键”排序，而我想根据“值”（计数）排序，所以指定key=第二元素
sorted(b.items(),key=lambda kv: -kv[1])
```

我们获取到整个数据集上的各个字符间二元组的计数。
对我们来说，将这些信息保存在一个二维数组（2-dimensional array）中而不是python字典会方便很多。行表示二元组的首字符，列表示二元组的第二个字符，二维数组的每个条目表示首的第二个字符跟随出现的频次计数
**特别的，我们将使用的数组表示或库是PyTorch**，PyTorch是一个深度学习神经网络框架，它的一部分包括torch.tensor，这使我们能够创建多维数组并对其进行非常高效的操作

```python
import torch
```

```python
N = torch.zeros((28, 28),dtype=torch.int32)
```

问题：有了这些字符串，我们现在需要对一个数组进行索引，并且使用整数进行索引，因此我们需要某种从字符到整数的查找表。
构建一个字符串数组，将所有的单词，全部连接成一个巨大的字符串。将这个传递给集合构造函数，它用于接收这个巨大的字符串，并去除重复项（集合不允许重复）

```python
chars = sorted(list(set(''.join(words))))
stoi = {s:i for i,s in enumerate(chars)}
stoi['<S>'] = 26
stoi['<E>'] = 27
```

```python
for w in words:
    chs = ['<S>'] + list(w) + ['<E>']
    for ch1, ch2 in zip(chs, chs[1:]):
        ix1 = stoi[ch1]
        ix2 = stoi[ch2]
        N[ix1, ix2] += 1
```

```python
#反转stoi即可
itos = {i:s for s,i in stoi.items()}
```

根据图反映的问题，进行改正：
* 张量换用27x27
* stoi、itos将首尾标记符号换用'.'，使得图更加美观

```python
N = torch.zeros((27, 27), dtype=torch.int32)
```

```python
chars = sorted(list(set(''.join(words))))
stoi = {s:i+1 for i,s in enumerate(chars)}
stoi['.'] = 0
itos = {i:s for s,i in stoi.items()}
```

```python
for w in words:
    chs = ['.'] + list(w) + ['.']
    for ch1, ch2 in zip(chs, chs[1:]):
        ix1 = stoi[ch1]
        ix2 = stoi[ch2]
        N[ix1, ix2] += 1
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(16,16))
plt.imshow(N, cmap='Blues')
for i in range(27):
    for j in range(27):
        chstr = itos[i] + itos[j] 
        plt.text(j, i, chstr, ha="center", va="bottom", color="gray")
        plt.text(j, i, N[i, j].item(), ha="center", va="top", color="gray")
plt.axis('off');
```

这个张量包含了从该二元字符级语言模型中采样所需的所有信息，接下来就是根据这些概率和计数，遵循采样信息


```python
N[0, :]
```

```python
#对计数进行归一化准备
p = N[0].float()

#方便概率运算
p = p/p.sum()
p
```

============================分界线=============================

为了统一输出，andrej在这里使用了生成器，手动约定随机数种子，从而生成相同的随机数

torch.multinomial方法可以根据给定权重对数组进行多次采样，返回采样后的**元素下标**(ix)
1. 第一个参数代表权重，即取每个值的概率
2. num_sample代表“采样次数”
3. replacement：默认值false，表示不放回采样

```python
g = torch.Generator().manual_seed(2147483647)
ix = torch.multinomial(p, num_samples=1, replacement=True, generator=g).item()
itos[ix]
```

通过加入约定的随机数种子了的**生成器**（generator）返回的生成器对象g，并用**torch.rand**方法随机生成**一个形状（shape）为1x3的张量**，注意rand函数用于生成具有**均匀分布**的随机数，范围在0~1之间

```python
g = torch.Generator().manual_seed(2147483647)
p = torch.rand(3, generator=g)
#归一化：
p = p/p.sum()
p
```

```python
index = torch.multinomial(p, num_samples=1, replacement=True, generator=g).item()
index
```

```python
g = torch.Generator().manual_seed(2147483647)

for i in range(20):
    out = []
    ix = 0
    while True:
        p = N[ix].float()
        p = p/p.sum()
        
        #使p为值1的27个元素，目的是均匀分布各元素可能性相等
        #p = torch.ones(27) / 27.0
        
        ix = torch.multinomial(p, num_samples=1, replacement=True, generator=g).item()
        out.append(itos[ix])
        if ix == 0:
            break
    print(''.join(out))
```

可以看到Bigram的效果不是很好，可能原因是：
1. 二元字符组模型本身的局限性

解决一个效率问题：

循环体中矩阵数据类型转换为浮点数，以及归一化操作，是非常低效的

```python
P = N.float()
#仅对行求和。压缩0维度（纵向压缩）：
#P.sum(0, keepdim=True)
#P.sum(0, keepdim=True).shape
#这里的除法使用了广播机制
P = P / P.sum(1, keepdim=True)
```

```python
g = torch.Generator().manual_seed(2147483647)

for i in range(20):
    out = []
    ix = 0
    while True:
        p = P[ix]
        ix = torch.multinomial(p, num_samples=1, replacement=True, generator=g).item()
        out.append(itos[ix])
        if ix == 0:
            break
    print(''.join(out))
```
